```python
import os
import cv2
def create_formated_label(in_label_dir, out_label_dir):
  names= os.listdir(in_label_dir)
  names.sort()
  paths = [os.path.join(in_label_dir, name) for name in names]
  for i,path in enumerate(paths):
    out_name = os.path.splitext(names[i])[0] + ".png"
    out_path = os.path.join(out_label_dir, out_name)
    img = cv2.imread(path, 0)
    if i%20 == 0:
      print("\n{}: ".format(i), path)
      print(out_path)
    img[img<15]=0
    img[img>=15]=1
    cv2.imwrite(out_path, img)
    # break
```
